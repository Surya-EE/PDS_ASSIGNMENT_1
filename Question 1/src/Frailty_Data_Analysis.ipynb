```python
import pandas as pd

# Load the cleaned dataset
file_path = 'Frailty_cleaned_data.csv'
data = pd.read_csv(file_path)

data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

# Boxplot for grip strength based on frailty status

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='frailty', y='Grip strength', data=data, palette='Set2')
plt.title('Grip Strength by Frailty Status')
plt.xlabel('Frailty Status')
plt.ylabel('Grip Strength (kg)')
plt.show()

plt.savefig('grip_strength_by_frailty_status.png', bbox_inches='tight')
plt.show()
```

# Countplot for Frailty status

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='frailty', data=data, palette='Set2')
plt.title('Count of Frailty Status')
plt.xlabel('Frailty')
plt.ylabel('Count')
plt.show()


plt.savefig('count_of_frailty_status.png', bbox_inches='tight')
plt.show()
```

# Scatter plot for Age vs Grip Strength

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Grip strength', hue='frailty', style='frailty', data=data, palette='Set2', s=100)
plt.title('Age vs Grip Strength')
plt.xlabel('Age (years)')
plt.ylabel('Grip Strength (kg)')
plt.legend(title='Frailty Status')
plt.show()

plt.savefig('age_vs_grip_strength.png', bbox_inches='tight')
plt.show()
```
